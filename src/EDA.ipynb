```python
import matplotlib.pyplot as plt

hha_yes = df[df['HadHeartAttack'] == 'Yes']['BMI']
hha_no = df[df['HadHeartAttack'] == 'No']['BMI']

hha_yes_mean = hha_yes.mean()
hha_no_mean = hha_no.mean()
hha_yes_median = hha_yes.median()
hha_no_median = hha_no.median()

plt.figure(figsize=(7, 5))

# Create the violin plot
parts = plt.violinplot([hha_no, hha_yes], showmeans=False, showextrema=True, showmedians=True)

# Customize the appearance of the violins
for i, pc in enumerate(parts['bodies']):
    pc.set_facecolor(['lightblue', 'lightcoral'][i])  # Different colors for 'No' and 'Yes'
    pc.set_edgecolor(['blue', 'red'][i])  # Matching outline colors
    pc.set_alpha(0.8)

# Customize median line appearance
parts['cmedians'].set_linewidth(1.5)

# Add titles and labels
plt.xlabel('Had Heart Attack', fontsize=14)
plt.ylabel('BMI (kg/m²)', fontsize=14)
plt.xticks([1, 2], ['No', 'Yes'], fontsize=12)

# Add median values as text annotations
plt.text(1, hha_no_median + 0.5, f"Median: {hha_no_median:.2f}", ha='center', fontsize=12, color='blue')
plt.text(2, hha_yes_median + 0.5, f"Median: {hha_yes_median:.2f}", ha='center', fontsize=12, color='red')

# Add grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Adjust layout and show the plot

plt.tight_layout()
plt.savefig('Fig1.jpeg')
plt.show()




# Calculate and print the averages for BMI based on Heart Attack occurrence
hha_yes_mean = hha_yes.mean()
hha_no_mean = hha_no.mean()

# Calculate the median for BMI based on Heart Attack occurrence
hha_yes_median = hha_yes.median()
hha_no_median = hha_no.median()

# Print averages and medians with units
print("Average BMI for individuals who had a heart attack:", f"{hha_yes_mean:.2f} kg/m²")
print("Median BMI for individuals who had a heart attack:", f"{hha_yes_median:.2f} kg/m²")
print("Average BMI for individuals who did not have a heart attack:", f"{hha_no_mean:.2f} kg/m²")
print("Median BMI for individuals who did not have a heart attack:", f"{hha_no_median:.2f} kg/m²")
```

```python
from sklearn.model_selection import train_test_split

y = df['HadHeartAttack']
X = df.loc[:, df.columns != 'HadHeartAttack']

random_state = 42

X_train, X_other, y_train, y_other = train_test_split(X,y,train_size = 0.6,stratify=y,random_state=random_state)
X_val, X_test, y_val, y_test = train_test_split(X_other,y_other,train_size = 0.5,stratify=y_other,random_state=random_state)

print(np.unique(y_train,return_counts=True))
print(np.unique(y_val,return_counts=True))
print(np.unique(y_test,return_counts=True))

print('Training set: ' , X_train.shape)
print('Validation set: ' , X_val.shape)
print('Test set: ' , X_test.shape)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split

onehot_ftrs = ['State', 'Sex', 'PhysicalActivities', 'HadAngina', 'HadStroke', 'HadAsthma', 'HadSkinCancer', 'HadCOPD', 'HadDepressiveDisorder', 'HadKidneyDisease', 'HadArthritis', 'HadDiabetes', 'DeafOrHardOfHearing', 'BlindOrVisionDifficulty', 'DifficultyConcentrating', 'DifficultyWalking', 'DifficultyDressingBathing', 'DifficultyErrands', 'ChestScan', 'RaceEthnicityCategory', 'AlcoholDrinkers', 'HIVTesting', 'FluVaxLast12', 'PneumoVaxEver', 'TetanusLast10Tdap', 'HighRiskLastYear', 'CovidPos']
ordinal_ftrs = ['GeneralHealth', 'LastCheckupTime', 'RemovedTeeth', 'SmokerStatus', 'ECigaretteUsage', 'AgeCategory']
ordinal_cats = [['Poor', 'Fair', 'Good', 'Very good', 'Excellent'], ['5 or more years ago', 'Within past 5 years (2 years but less than 5 years ago)', 'Within past 2 years (1 year but less than 2 years ago)', 'Within past year (anytime less than 12 months ago)'], ['All', '6 or more, but not all', '1 to 5', 'None of them'], ['Current smoker - now smokes every day', 'Current smoker - now smokes some days', 'Former smoker', 'Never smoked'], ['Use them every day', 'Use them some days', 'Not at all (right now)', 'Never used e-cigarettes in my entire life'], ['Age 18 to 24', 'Age 25 to 29', 'Age 30 to 34', 'Age 35 to 39', 'Age 40 to 44', 'Age 45 to 49', 'Age 50 to 54', 'Age 55 to 59', 'Age 60 to 64', 'Age 65 to 69', 'Age 70 to 74', 'Age 75 to 79', 'Age 80 or older']]
minmax_ftrs = ['PhysicalHealthDays', 'MentalHealthDays', 'SleepHours', 'HeightInMeters', 'WeightInKilograms', 'BMI']
print (len(onehot_ftrs))
print (len(ordinal_ftrs))
print (len(minmax_ftrs))

preprocessor = ColumnTransformer(
    transformers=[
        ('ord', OrdinalEncoder(categories = ordinal_cats), ordinal_ftrs),
        ('onehot', OneHotEncoder(sparse_output=False,handle_unknown='ignore'), onehot_ftrs),
        ('minmax', MinMaxScaler(), minmax_ftrs)])

clf = Pipeline(steps=[('preprocessor', preprocessor)])

X_train_prep = clf.fit_transform(X_train)
X_val_prep = clf.transform(X_val)
X_test_prep = clf.transform(X_test)

print(X_train.shape)
print(X_train_prep.shape)
print(X_train_prep)

y_train_encoded = y_train.map({'No': 0, 'Yes': 1})
y_val_encoded = y_val.map({'No': 0, 'Yes': 1})
y_test_encoded = y_test.map({'No': 0, 'Yes': 1})

print(y_train_encoded.shape)
print(y_val_encoded.shape)
print(y_test_encoded.shape)
```
